1. 다음과 같은 데이터프레임을 생성한다.

```
import pandas as pd
import numpy as np 
df = pd.DataFrame(dict(Gender=['M', 'F', 'M', 'F',
                               'F', 'M', 'M', np.nan],
                       State=['NY','IL', 'NY', 'CA',
                              'IL', 'CA', 'CA', 'IL'],
                       col_1=[10,20,30,np.nan,40,50,60,70],
                       col_2=[1,6,2,4,6,9,np.nan,3]))
```

````
import pandas as pd
import numpy as np 
df = pd.DataFrame(dict(Gender=['M', 'F', 'M', 'F',
                               'F', 'M', 'M', np.nan],
                       State=['NY','IL', 'NY', 'CA',
                              'IL', 'CA', 'CA', 'IL'],
                       col_1=[10,20,30,np.nan,40,50,60,70],
                       col_2=[1,6,2,4,6,9,np.nan,3]))

df
````

2. 데이터프레임에서 'Gender'별 평균을 구하시오.

````
df.groupby('Gender').mean()
````

3. 데이터프레임에서 'Gender'별, 'State'별 평균을 구하시오.

````
df.groupby(['Gender','State']).mean()
````

4. 데이터프레임에서 'Gender'별, 'State'별 각 column의 최소, 최대 값을 구하시오.

````
df.groupby(['Gender','State']).agg([min, max])
````

5. 다음과 같은 데이터프레임을 생성하시오.
```
df1 = pd.DataFrame({
    'ID': [1001, 1002, 1003, 1004, 1005, 1006, 1007],
    'Name': ['Jane', 'James', 'Michel', 'Sam', 'Suji', 'Kurt', 'Julie']
})
df2 = pd.DataFrame({
    'ID': [1001, 1001, 1005, 1006, 1008, 1001],
    'Deposit': [10000, 20000, 15000, 5000, 100000, 30000]
})
```

````
df1 = pd.DataFrame({
    'ID': [1001, 1002, 1003, 1004, 1005, 1006, 1007],
    'Name': ['Jane', 'James', 'Michel', 'Sam', 'Suji', 'Kurt', 'Julie']
})

df1
````

````
df2 = pd.DataFrame({
    'ID': [1001, 1001, 1005, 1006, 1008, 1001],
    'Deposit': [10000, 20000, 15000, 5000, 100000, 30000]
})
df2
````

6. merge 명령으로 위의 두 데이터프레임 df1, df2 를 수평으로 합쳐 출력하시오. ID 컬럼과 NO 컬럼을 기준으로 inner join을 수행하시오.


````
pd.merge(df1, df2)
````

7. merge를 사용하여 df1, df2 를 join 하는데 full outer join으로 수행하시오.

````
pd.merge(df1, df2, how='outer')
````

8. merge 대신에 join 명령어를 사용하여 full outer join을 수행하시오.

````
df1.set_index('ID').join(df2.set_index('ID'), how='outer')
````
